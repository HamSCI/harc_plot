```python
import aacgmv2
import datetime
import numpy as np
import pandas as pd
```

```python
dtime = datetime.datetime(2017,9,1)

fname = 'kp_stations.csv'
df    = pd.read_csv(fname)

df_lst = []
for rinx,row in df.iterrows():
    aa = aacgmv2.get_aacgm_coord(row['glat'],row['glon'],0,dtime)
    row['mlat'] = aa[0]
    row['mlon'] = aa[1]
    row['mlt']  = aa[2]
    df_lst.append(row)
    
df = pd.DataFrame(df_lst)
```
